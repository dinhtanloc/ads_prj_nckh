- POLICY_EFF_DT: Policy Effective Date (ngày chính sách hiệu lực)
- LOSS_DT: Date on which loss happened (ngày diễn ra mất mát)
- REPORT_DT: Date on which the incident was reported (ngày tai nạn được báo cáo)

```python
# Xem biểu diễn sơ qua trước
print(df[['DATE_OF_JOINING', 'TXN_DATE_TIME', 'POLICY_EFF_DT', 'LOSS_DT', 'REPORT_DT']].to_string(index=False))
```

Đầu tiên, cần kiểm tra tính đúng đắn trước:
- DATE_OF_JOINING phải diễn ra trước TXN_DATE_TIME
- LOSS_DT phải diễn ra trước hoặc cùng ngày REPORT_DT
- REPORT_DT phải diễn ra cùng ngày hoặc trước TXN_DATE_TIME
- POLICY_EFF_DTT nên diễn ra trước TXN_DATE_TIME (có thể có trường hợp diễn ra vấn đề trong các giao dịch trước, nhưng sau này chính sách mới cập nhật những điều luật).

```python
is_ordered = (df['LOSS_DT'] <= df['REPORT_DT']).all()

if is_ordered:
    print("LOSS_DT diễn ra trước REPORT_DT trong tất cả các hàng.")
else:
    print("LOSS_DT không diễn ra trước REPORT_DT trong ít nhất một hàng.")
```

```python
is_ordered = (df['REPORT_DT'] <= df['TXN_DATE_TIME']).all()

if is_ordered:
    print("REPORT_DT diễn ra trước TXN_DATE_TIME trong tất cả các hàng.")
else:
    print("REPORT_DT không diễn ra trước TXN_DATE_TIME trong ít nhất một hàng.")
```

```python
is_ordered = (df['POLICY_EFF_DT'] <= df['TXN_DATE_TIME']).all()
if is_ordered:
    print("POLICY_EFF_DT diễn ra trước TXN_DATE_TIME trong tất cả các hàng.")
else:
    print("POLICY_EFF_DT không diễn ra trước TXN_DATE_TIME trong ít nhất một hàng.")
```

Dữ liệu đã đúng hết. Bỏ biến policy vì chưa cần dùng đến trong phạm vi bào nghiên cứu

Thực hiện:
- Lấy ngày giao dịch trừ cho ngày vào làm để tạo ra biến năm kinh nghiệm của nhân viên là EXP_YEAR.
- Lấy ngày report trừ ngày xảy ra tai nạn để tạo ra biến tốc độ ngày báo cáo kể từ tai nạn là DELAY_DT
- Lấy ngày transaction trừ ngày report để tạo ra biến tốc độ ngày thực hiện đơn của bên bảo hiểm là WAIT_DT

* Các biến mới tạo đều phải là giá trị int64

Xóa các biến cũ khi đã tạo hết các biến mới



```python
main.drop('POLICY_EFF_DT',axis=1,inplace =True)
```

```python
# Chuyển đổi cột 'DATE_OF_JOINING' và 'TXN_DATE_TIME' thành kiểu dữ liệu datetime
main['DATE_OF_JOINING'] = pd.to_datetime(main['DATE_OF_JOINING'])
main['TXN_DATE_TIME'] = pd.to_datetime(main['TXN_DATE_TIME'])

# Tính số năm kinh nghiệm và tạo cột 'EXP_YEAR'
main['EXP_YEAR'] = (main['TXN_DATE_TIME'] - main['DATE_OF_JOINING']).dt.days // 365

# Đổi kiểu dữ liệu của cột 'EXP_YEAR' thành int64
#main['EXP_YEAR'] = main['EXP_YEAR'].astype('int64')

# Replace non-finite values with a default value (e.g., 0)
main['EXP_YEAR'] = main['EXP_YEAR'].fillna(0)

# Convert the column to int64
main['EXP_YEAR'] = main['EXP_YEAR'].astype('int64')

# In kết quả
print(main['EXP_YEAR'])
```

```python
main['REPORT_DT'] = pd.to_datetime(main['REPORT_DT'])
main['LOSS_DT'] = pd.to_datetime(main['LOSS_DT'])

# Tính số ngày chênh lệch và lưu vào cột 'DELAY_DT'
main['DELAY_DT'] = (main['REPORT_DT'] - main['LOSS_DT']).dt.days

print(main['DELAY_DT'])
```

```python
main['TXN_DATE_TIME'] = pd.to_datetime(main['TXN_DATE_TIME'])
main['REPORT_DT'] = pd.to_datetime(main['REPORT_DT'])

# Tính số ngày chênh lệch và lưu vào cột 'DELAY_DT'
main['WAIT_DT'] = (main['TXN_DATE_TIME'] - main['REPORT_DT']).dt.days

print(main['WAIT_DT'])
```

```python
main.drop('DATE_OF_JOINING',axis=1,inplace =True)
main.drop('TXN_DATE_TIME',axis=1,inplace =True)
main.drop('LOSS_DT',axis=1,inplace =True)
main.drop('REPORT_DT',axis=1,inplace =True)
```
